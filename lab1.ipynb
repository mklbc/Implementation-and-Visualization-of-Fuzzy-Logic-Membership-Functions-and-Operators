```python
import numpy as np
import matplotlib.pyplot as plt

# Define a range for the universe of discourse
x = np.linspace(0, 10, 100)

# Task 1: Triangular and Trapezoidal Membership Functions
def triangular_mf(x, a, b, c):
    return np.maximum(0, np.minimum((x - a) / (b - a), (c - x) / (c - b)))

def trapezoidal_mf(x, a, b, c, d):
    return np.maximum(0, np.minimum(np.minimum((x - a) / (b - a), 1), (d - x) / (d - c)))

# Task 2: Gaussian Membership Function (Two-Sided)
def gaussian_mf(x, mean, std_dev):
    return np.exp(-((x - mean) ** 2) / (2 * std_dev ** 2))

# Task 3: Generalized Bell Membership Function
def generalized_bell_mf(x, a, b, c):
    return 1 / (1 + np.abs((x - c) / a) ** (2 * b))

# Task 4: Set of Sigmoid Functions
def one_sided_sigmoid(x, a, b):
    return 1 / (1 + np.exp(-a * (x - b)))

def two_sided_sigmoid(x, a, b):
    return 1 / (1 + np.exp(-a * (x - b))) - 1 / (1 + np.exp(a * b))

def asymmetric_sigmoid(x, a, b, c):
    return 1 / (1 + np.exp(-a * (x - b))) - 1 / (1 + np.exp(a * (x - c)))

# Task 5: Set of Polynomial Membership Functions (Z-, PI-, and S-functions)
def z_shaped_polynomial(x, a, b, c):
    return np.piecewise(x, [x < a, (a <= x) & (x <= b), x > c], [0, lambda x: 2 * (x - a) ** 2 / ((b - a) * (c - a)), 1])

def pi_shaped_polynomial(x, a, b, c):
    return np.piecewise(x, [x < a, (a <= x) & (x <= b), (b <= x) & (x <= c), x > c], [0, lambda x: 2 * ((x - a) / (b - a)) ** 2, lambda x: 1 - 2 * ((x - b) / (c - b)) ** 2, 0])

def s_shaped_polynomial(x, a, b, c):
    return np.piecewise(x, [x < a, (a <= x) & (x <= b), x > c], [0, lambda x: 2 * ((x - a) / (b - a)) ** 2, 1])

# Task 6: Minimax Interpretation of Logical Operators
def logical_and_minimax(inputs):
    return 1 if min(inputs) == 1 else 0

def logical_or_minimax(inputs):
    return 0 if max(inputs) == 0 else 1

# Task 7: Plausible Interpretation of Conjunctive and Disjunctive Operators
def conjunctive_operator(a, b):
    return min(a, b)

def disjunctive_operator(a, b):
    return max(a, b)

# Task 8: Complement of a Fuzzy Set (Not used temperature as requested)
def fuzzy_set_complement(x):
    return 1 - x

# Example usage of the functions
if __name__ == "__main__":
    # Task 1: Triangular and Trapezoidal Membership Functions
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    plt.plot(x, triangular_mf(x, 2, 4, 7), label='Triangular')
    plt.xlabel('Input')
    plt.ylabel('Membership Degree')
    plt.title('Triangular Membership Function')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(x, trapezoidal_mf(x, 2, 3, 7, 8), label='Trapezoidal')
    plt.xlabel('Input')
    plt.ylabel('Membership Degree')
    plt.title('Trapezoidal Membership Function')
    plt.legend()

    plt.tight_layout()

    # Task 2: Gaussian Membership Function (Two-Sided)
    plt.figure(figsize=(6, 4))

    plt.plot(x, gaussian_mf(x, 5, 1), label='Gaussian')
    plt.xlabel('Input')
    plt.ylabel('Membership Degree')
    plt.title('Gaussian Membership Function')
    plt.legend()

    # Task 3: Generalized Bell Membership Function
    plt.figure(figsize=(6, 4))

    plt.plot(x, generalized_bell_mf(x, 2, 2, 4), label='Generalized Bell')
    plt.xlabel('Input')
    plt.ylabel('Membership Degree')
    plt.title('Generalized Bell Membership Function')
    plt.legend()

    # Task 4: Set of Sigmoid Functions
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 3, 1)
    plt.plot(x, one_sided_sigmoid(x, 2, 2), label='One-Sided')
    plt.xlabel('Input')
    plt.ylabel('Membership Degree')
    plt.title('One-Sided Sigmoid Function')
    plt.legend()

    plt.subplot(1, 3, 2)
    plt.plot(x, two_sided_sigmoid(x, 2, 2), label='Two-Sided')
    plt.xlabel('Input')
    plt.ylabel('Membership Degree')
    plt.title('Two-Sided Sigmoid Function')
    plt.legend()

    plt.subplot(1, 3, 3)
    plt.plot(x, asymmetric_sigmoid(x, 2, 2, 6), label='Asymmetric')
    plt.xlabel('Input')
    plt.ylabel('Membership Degree')
    plt.title('Asymmetric Sigmoid Function')
    plt.legend()

    plt.tight_layout()

    # Task 5: Set of Polynomial Membership Functions (Z-, PI-, and S-functions)
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 3, 1)
    plt.plot(x, z_shaped_polynomial(x, 2, 3, 5), label='Z-shaped')
    plt.xlabel('Input')
    plt.ylabel('Membership Degree')
    plt.title('Z-shaped Polynomial Membership Function')
    plt.legend()

    plt.subplot(1, 3, 2)
    plt.plot(x, pi_shaped_polynomial(x, 2, 3, 5), label='Pi-shaped')
    plt.xlabel('Input')
    plt.ylabel('Membership Degree')
    plt.title('Pi-shaped Polynomial Membership Function')
    plt.legend()

    plt.subplot(1, 3, 3)
    plt.plot(x, s_shaped_polynomial(x, 2, 3, 5), label='S-shaped')
    plt.xlabel('Input')
    plt.ylabel('Membership Degree')
    plt.title('S-shaped Polynomial Membership Function')
    plt.legend()

    plt.tight_layout()

    # Task 6: Minimax Interpretation of Logical Operators
    input_set1 = [1, 0, 1]
    input_set2 = [0, 0, 0]
    input_set3 = [1, 1, 1]

    result_and = logical_and_minimax(input_set1)
    result_or = logical_or_minimax(input_set2)

    print(f"Logical AND result: {result_and}")
    print(f"Logical OR result: {result_or}")

    # Task 7: Plausible Interpretation of Conjunctive and Disjunctive Operators
    conj_result = conjunctive_operator(1, 0)
    disj_result = disjunctive_operator(1, 0)

    print(f"Conjunctive Operator Result: {conj_result}")
    print(f"Disjunctive Operator Result: {disj_result}")

    # Task 8: Complement of a Fuzzy Set
    plt.figure(figsize=(6, 4))

    original_set = np.array([0.2, 0.4, 0.6, 0.8])
    complement_set = fuzzy_set_complement(original_set)

    plt.plot(original_set, label='Original Set')
    plt.plot(complement_set, label='Complement Set')
    plt.xlabel('Elements')
    plt.ylabel('Membership Degree')
    plt.title('Complement of a Fuzzy Set')
    plt.legend()

    plt.show()
```
